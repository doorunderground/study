```python
a = np.zeros((4,5), np.uint8)
a[1,2] = 255
plt.imshow(a, cmap='gray')

a[:,1] = 255
plt.imshow(a, cmap='gray')

a[2,:] = 128
plt.imshow(a, cmap='gray')


a[3,[0,2,3]] = 64
plt.imshow(a, cmap='gray')
```
