# 두 점을 지나는 직선의 방정식을 구하시오!

[ x1  1 ] [a] = [y1]
[ x2  1 ] [b] = [y2]

```python
import numpy as np

x = np.array([1,2]).reshape(2, 1)
y = np.array([3,5]).reshape(-1, 1)

# y1 = ax1 + b
# y2 = ax2 + b

# 2,1는 사람이 직접 맞춰주는거고, -1, 1은 numpy가 알아서 맞춰라 느낌?

xb = np.c_[x, np.ones((2,1))]

# [1  2  1]
# [3  5  1]

theta = np.linalg.inv(xb) @ y
#행렬 곱은 '@' 로 표시

a = theta[0, 0]
b = theta[1, 0]

a,b
```

# 점이 3개 이상일 때 직선을 어떻게 구하냐!

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3]).reshape(-1,1)
y = np.array([4, 5.1, 5.8]).reshape(-1,1)

xb = np.c_[x, x**0]

#[x1  1]           [y1]
#[x2  1]  [a]   =  [y2]  
#[x3  1]  [b]      [y3]

xbpi = np.linalg.inv(xb.T @ xb) @ xb.T
theta = xbpi @ y

a = theta[0,0]
b = theta[1,0]

a,b
```

```python
def func(x):
    return x * a + b

xt = np.array([1,3])
yt = func(xt)

plt.plot(x, y, '.')
plt.plot(xt,yt, '-')
#완벽하게 지나는 직선을 구하진 못하고, 가장 최적의 직선을 구하게 됨
```

# 100개의 점을 지나는 직선의 방정식을 구하라?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(100, 1) * 2
y = x * 3 + 4 + np.random.randn(100, 1)
plt.plot(x, y, '.')
```

```python
xb = np.c_[x, x**0]
theta = np.linalg.pinv(xb) @ y

a, b = theta.reshape(-1)

def func(x):
    return x * a + b

xt = np.array([0,2])
yt = func(xt)

plt.plot(x, y, '.')
plt.plot(xt, yt, '-')
```

# 2차 방정식 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x.... -3 ~ +3 

x = np.random.rand(100, 1) * 6 - 3
y = 0.5 * x**2 + x + 2 + np.random.randn(100, 1)


plt.plot(x,y,'.') 
```

```python
#1
xb = np.c_[x**2, x,  x**0]

theta = np.linalg.pinv(xb) @ y

a, b, c = theta.reshape(-1)

def func(x):
    return a*x**2 + b*x + c

xt = np.linspace(-3, 3, 100).reshape(-1,1)
yt = func(xt)

plt.plot(x, y, '.')
plt.plot(xt, yt, '-')
```

```python
#2
from sklearn.linear_model import LinearRegression

lin = LinearRegression()
lin.fit(x, y)

xt = np.linspace(-3, 3, 100).reshape(-1,1)
yt = lin.predict(xt)

plt.plot(x, y, '.')
plt.plot(xt, yt, '-')
```

# 해석적 방법은 선형방정식에서만 되는데.... 곡선의 방정식을 풀어벼렸네?

# 선형방정식
 - 직선/평면 선형, 곡선/곡면 비선형
 - 행렬곱으로 표현 가능하면 선형, 그렇지 않으면 비선형

```python
# x → [1, x, x², x³, …, x¹⁰] 형태로 변환하는 함수
#머신러닝에서 Polynomial Feature(다항 특징) 만들 때 쓰는 전형적인 코드

#커널 만드는?
x = np.random.rand(1000, 1) * 6 - 3
y = 0.5 * x**2 + x + 2 + np.random.randn(1000, 1)

n = 50

def kernel(x):
    poly_list = []
    for i in range(n + 1):
        poly_list.append(x ** i)
    return np.concat(poly_list, axis=1)

xb = kernel(x)
M = xb[:,1:].mean(axis=0)
S = xb[:,1:].std(axis=0)
xb[:,1:] -= M
xb[:,1:] /= S

theta = np.linalg.pinv(xb) @ y

xt = np.linspace(-3, 3, 1000).reshape(-1, 1)
xtb = kernel(x)
xtb[:,1:] -= M
xtb[:,1:] /= S
yt = xtb @ theta

plt.axis([-3, 3, 0, 10])
plt.plot(x, y, '.')
plt.plot(xt, yt, '-')
```

# scikit-learn 해보기

```python
#다항식 kernel? 
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 10, include_bias=False)
#xt = np.array([1.,2.,3.]).reshape(-1, 1)
poly.fit(x) # 내부에 컬럼이 몇 개 인지 파악
xp = poly.transform(x)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(xp)
xps = scaler.transform(xp)

from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(xps, y)

#test
xt = np.linspace(-3, 3, 100).reshape(-1, 1)
xtp = poly.transform(xt)
xtps = scaler.transform(xtp)
yt = lin.predict(xtps)

plt.plot(x,y,'.')
plt.plot(xt,yt,'-')
```

```python
from sklearn.pipeline import Pipeline

model = Pipeline([
    ('poly', PolynomialFeatures(degree = 10, include_bias=False)),
    ('scaler', StandardScaler()),
    ('lin', LinearRegression())
])

model.fit(x,y)

xt = np.linspace(-3, 3, 100).reshape(-1, 1)
yt = model.predict(xt)

plt.plot(x,y,'.')
plt.plot(xt,yt,'-')
```
