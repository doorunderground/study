```python
a = np.arange(6).reshape(3,2,1) #2x1 3겹
b = a[0]    
c = a[:2, 0]
d = a[0, :, 0]

print(b)
print(c)
print(d)
```

a =
[
  [ [0],
    [1] ],   ← a[0]

  [ [2],
    [3] ],   ← a[1]

  [ [4],
    [5] ]    ← a[2]
]

c = a[:2, 0] -> a[0],a[1] 겹 + 각 겹의 0번째 행
d = a[0, : , 0] -> a[0]겹 + 각 겹의 전체 + 값만 꺼내기

```python
a = np.arange(24).reshape(2, 4, 1, 3) #4x1x3 2겹
# 1
b = a[:, 0]
c = a[0, :, :1, 0]  # 겹 줄 세로 채널
d = a[1, 1, 0]
e = a[:, 1:, 0, 2]

print(b)
print()

print(c)
print()

print(d)
print()

print(e)
print()
```
