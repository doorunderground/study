# 1
Pandas 함수로 2개 데이터 파일을 읽고 합쳐서 1개의 데이터프레임 변수명 df에 할당하는 코드를 작성하세요.

LifeExpA.csv파일을 읽어 데이터프레임 변수명 df_a에 할당하세요.
LifeExpB 파일을 읽어 데이터 프레임 변수명 df_b에 할당하세요.
df_a, df_b를 합쳐서 하나의 데이터프레임으로 만드세요. 판다스의 concat을 활용하세요
합친 데이터셋의 index를 초기화하세요z

reset_index 활용
기존 인덱스 버림(drop) : True
처리 결과를 df에 저장하세요.

```python
import pandas as pd

df_a = pd.read_csv('C:\\Pandas_study\\pandas\\LifeExpA.csv')
df_b = pd.read_csv('C:\\Pandas_study\\pandas\\LifeExpB.csv')

#컬럼 'BMI ' -> 컬럼에 공백이 있음 그거 없애줌
#str.strip() -> 문자열로 다 바뀐뒤에 -> 문자열 및 공백 제거 
df_a.columns = df_a.columns.str.strip()
df_a.columns

df_b.columns = df_b.columns.str.strip()
df_a.columns
```

```python
df = pd.concat([df_a, df_b])
df
```

```python
#합치면서 정렬이 안되어있음
df = df.reset_index(drop=True)
df
```

# 5
데이터를 입력 X와 타깃 y로 분리하시오.
y : Life expectancy
X : Life expectancy 외 컬럼
sklearn의 train_test_split 함수를 활용하여 학습데이터와 테스트 데이터로 분리하시오.
random_state : 42
학습데이터 80%, 훈련데이터 20%


RobustScaler를 활용하여 입력 데이터 X_train, X_test에 대한 스케일링을 진행하시오.
X_train : fit_transform 활용
X_test : transform 활용

```python
x = df4.drop(columns = ['Life expectancy']) # 정답을 맞히는 데 사용할 재료들 Life expectancy 제외
y = df4['Life expectancy']  # 모델이 예측해야 할 값


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
# 전체 데이터 중 80%는 학습용, 20%는 테스트용을 나눠라
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
X_train = scaler. fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
sns.scatterplot(x=X_train[:, 0], y=y_train)
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(X_train, y_train)
pred = lin.predict(X_test)
#(예측값 - 실제값)^2 
#mse = 예측이 틀린 정도를 숫자로 표현한 것 
# 0에 가까울수록 예측 잘함

lin_mse = ((pred - y_test) ** 2).mean()
```

```python
lin_mse
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
pred = rf.predict(X_test)
rf_mse = ((pred - y_test) ** 2).mean()
rf_mse
```
